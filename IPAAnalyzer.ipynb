```python
#this will compare two 1 D histograms i.e. reco v true offsets

import plotly.graph_objects as go
import plotly
import math
import pandas as pd
import plotly.express as px
import numpy as np
import matplotlib.pyplot as plt

x0 = pd.read_csv('SGDv0.csv')

fig, ax = plt.subplots(1,1)
fig = go.Figure()
ax, fig.add_trace(go.Histogram(x=x0['reco'])) #blue
ax, fig.add_trace(go.Histogram(x=x0['true'])) # red
labels = ['reco','true']
fig.update_layout(barmode='overlay')

fig.update_traces(opacity=0.6)

fig.show()
```

```python
# Plots the residual i.e. true - reco
import plotly
import math
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('SGDv0.csv')

fig = px.histogram(df, x=(df['Residual']),nbins=50)
mean = round((np.mean(df['Residual'])),5)
st = round((np.std(df['Residual'])),5)
plt.text(0.08,50, r'$\mu=$'+str(mean)+' $\sigma=$ '+str(st))
fig.show() 
```
